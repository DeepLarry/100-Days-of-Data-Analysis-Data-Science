```python
import sympy as sp

# Variable define karo
x = sp.Symbol('x')

# Function likho: f(x) = x^2 + 5x
f = x**2 + 5*x

# Derivative nikalo (Power Rule khud lagayega)
derivative = sp.diff(f, x)

print(f"Function: {f}")
print(f"Derivative: {derivative}") # Output: 2*x + 5
```

```python
import torch

# Ek variable lo aur 'requires_grad=True' karo (matlab iska derivative nikalna hai)
x = torch.tensor(3.0, requires_grad=True)

# Function define karo: y = x^2
y = x**2

# Step 1: Calculate Error (y)
# Step 2: Backpropagate (Calculus chalao)
y.backward()

# Ab dekho x ka derivative (dy/dx) kya aaya
print(f"x = 3 par x^2 ka slope hai: {x.grad}") # Output: 6.0 (kyunki 2*3 = 6)
```
