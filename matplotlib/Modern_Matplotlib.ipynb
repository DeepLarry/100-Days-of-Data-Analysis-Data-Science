```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,4,5,6,7,8,9,10]

y1=[np.sqrt(i) for i in x] #square root
y2=[i*2 for i in x] #double
y3=[i**2 for i in x] # square
y4=[i**3 for i in x] #cubes

fig,ax=plt.subplots(2,2,figsize=(7,7))    #figsize se hum hight & width dekhate hai cansvas ka aur ese ghata aur badha sakate hai
ax[0][0].plot(x,y1)
ax[0][0].set_title("Plot1 - Square root")
ax[0][0].set_xlabel("x_axis values")
ax[0][0].set_ylabel("y_axis values")



ax[0][1].plot(x,y2)
ax[0][1].set_title("Plot2 - Double")

ax[1][0].plot(x,y3)
ax[1][0].set_title("Plot3 - Square")

ax[1][1].plot(x,y4)
ax[1][1].set_title("Plot4 - Qubes")

fig.suptitle("Multiple Plots")
fig.tight_layout()
```
