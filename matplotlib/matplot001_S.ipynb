```python
#Real Data
# Oscar winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "Agro"
]
Year=[2008,2009,2010,2011,2012]
oscar_revenue = [1005,170,427,133,232] #In $M
#Plot
plt.plot(Year,oscar_revenue)
plt.title("Oscar Movies Revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue in $M")
```

```python
non_oscar_movies=["Slumdog Millionnaire" , "Avatar","Inception","Lincoln"]
non_oscar_years=[2008,2009,2010,2011,2012]
non_oscar_revenue=[378,2799,898,233,244] #in $M

plt.plot(Year,oscar_revenue,color="black",marker=">",linestyle="--", label="Oscar Movies")
plt.plot(Year,non_oscar_revenue,color="yellow",linewidth="2",marker="o",label="Non-Oscar Movies")

plt.title("oscar Movie Revenue vs Non-Oscar Movies Reven")

plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")
plt.legend()

plt.legend(loc="upper right")
```
