```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,4,5,6,7,8,9,10]

y1=[np.sqrt(i) for i in x] #square root
y2=[i*2 for i in x] #double
y3=[i**2 for i in x] # square
y4=[i**3 for i in x] #cubes

#In first graph we want to show them relation between x and y1 along with there square root similarly apply everyone.
#
#Ham chahte hai ki hamare pass 2 row and 2 column aa jaye jisame hum subplot kar sake jisake liye
#-------plt.subplot(2,2,1)-Row,Column,index------------


plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("plot1- square root")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("plot2- Double")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("plot3- square")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("plot4- Qubes")
plt.tight_layout()

#-------------plt.tight_layout()---------
#Yeh function plots ke beech ka spacing automatically adjust karta hai
#Taaki labels, titles, axes overlap na karein.
```
