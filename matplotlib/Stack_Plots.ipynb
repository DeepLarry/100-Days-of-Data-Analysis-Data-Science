```python
#Stack Plots
#A stack plot is a type of plot where multiple data series are stacked on top of each other.
#Each "layer" show the contribution of one category, and the top line show the cumulative total.
```

```python
#Stack Plots
import matplotlib.pyplot as plt
days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
direct = [50,60,70,80,90,100,10]
organic=[30,40,50,35,70,20,110]
social=[20,25,30,35,40,50,60]

#plt.pie([direct[0],organic[0],social[0]],labels=["Direct","Organic","Social"],autopct="%1.1f%%")

##      autopct ka matlab = auto percentage

################          autopct="%1.1f%%"           ####################

#  % → formatting start
#  1.1f → 1 decimal place tak float value
#  %% → actual % symbol print karne ke liye


#Note:   ham sabhi data ko ek sath show karane ke liye alaga alag likhnge nahi esake liye hum stack plot ka use karenge.
#Hum chahte hai ki x and y asix per hum work kare - jisake liye x-asix per hum day ko chahate hai aur y-axis per hum value chate hai sabhi ka ek sath to stack plot use karenge.
plt.stackplot(days,direct,organic,social,labels=["Direct","Organic","Social"])
plt.title("Marketing data for the week")
plt.xlabel("Days")
plt.ylabel("number of customers come")
plt.legend()
```
