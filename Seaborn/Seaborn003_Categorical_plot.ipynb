```python
#Seaborn
#Categorical Plot :-
        #In Seaborn, categorical plots are used when:
        #One variable = category (names, groups)
        #Other variable = numeric (values)
#Example:
#city vs temperature
#gender vs salary

#Bar plot :Compare average values across categories
#Shows mean (average) by default
#Automatically calculates values
#Can use hue for comparison


#Box plot: Show data distribution + outliers

#Median (middle line)
#Q1, Q3 (box)
#Outliers (dots)

#💡 Best for:
#👉 detecting outliers
#👉 understanding spread of data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')

sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python

#Bar Plots
tips = sns.load_dataset('tips')

sns.boxplot(
    data=tips,
    x="day",
    y="tip",
    # hue="sex"
)
```

```python
#Seaborn
#Distribution Plots: Used to understand how data is distributed
        #(a) Histogram (sns.histplot())
        #Show frequency of values -How many times values occur

        #(b) KDE Plot (sns.kdeplot())
        #Purpose:Smooth curve of distribution
```
