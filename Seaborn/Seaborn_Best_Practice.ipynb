```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips=sns.load_dataset("tips")
fig,ax=plt.subplots()
sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    marker="o",
    ax=ax,
   ci=None     #Confidence Interval- remove backpart interval img
)
ax.set_title("avg bill value each day")
ax.set_xlabel("Days")
ax.set_ylabel("Total Bill Amount")

fig.tight_layout()
```
