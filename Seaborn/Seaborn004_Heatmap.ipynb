```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

flights=sns.load_dataset("Flights")
flight.head()
```

```python
# flight.sort_values(by=["year","month"])
flight_pivot=flights.pivot(index="month",columns="year",values="passengers")
print(flight_pivot)

sns.heatmap(
    flight_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)
plt.title("passenger")
plt.tight_layout()
```
