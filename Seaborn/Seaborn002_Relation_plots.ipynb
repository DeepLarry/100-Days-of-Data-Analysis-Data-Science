```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
#Line plot

tips = sns.load_dataset('tips')

sns.lineplot(
    data=tips,
    x="total_bill",
    y="tip",
    # hue="time"
)
```

```python
flight=sns.load_dataset('flights')

sns.lineplot(
    data=flight,
    x="year",
    y="passengers"
)

```
