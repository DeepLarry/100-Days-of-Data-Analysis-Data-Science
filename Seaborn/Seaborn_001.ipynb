```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
sns.set_theme(style="dark")     #Set overall style/design of plots
                                #dark"
                                #"white"
                                #"darkgrid" ✅ (most used)
                                #"whitegrid" ✅ (best for beginners)

sns.relplot(             #Used to show relationship between two variables
    data=tips,
    x="total_bill",     #x and y define axes
    y="tip",
    kind="line",       #"scatter" default   #Line - line graph
    hue="smoker",    #Adds color grouping
    size="size",     #Changes size of lines/points
    # style="smoker"      #style- Changes line style / markers
)
```
